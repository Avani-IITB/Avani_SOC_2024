```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
sample_submission = pd.read_csv('sample_submission.csv')
```

```python
train.head()
```

```python
label_encoder = LabelEncoder()
```

```python
train['store'] = label_encoder.fit_transform(train['store'])
```

```python
train['item'] = label_encoder.fit_transform(train['item'])
```

```python
test['store'] = label_encoder.transform(test['store'])
```

```python
test['item'] = label_encoder.transform(test['item'])
```

```python
train['year'] = train.index.year
```

```python
train['month'] = train.index.month
```

```python
train['day'] = train.index.day
```

```python
train['dayofweek'] = train.index.dayofweek
```

```python
test['year'] = test.index.year
```

```python
test['month'] = test.index.month
```

```python
test['day'] = test.index.day
```

```python
test['dayofweek'] = test.index.dayofweek
```

```python
X = train[['store', 'item', 'year', 'month', 'day', 'dayofweek']]
y = train['sales']
```

```python
X_train, X_test = X.iloc[:-90], X.iloc[-90:]
y_train, y_test = y.iloc[:-90], y.iloc[-90:]
```

```python
model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=1000)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
print(f'Mean Squared Error: {mse}') #again, damnnn!
```

```python
test_features = test[['store', 'item', 'year', 'month', 'day', 'dayofweek']]
```

```python
test_pred = model.predict(test_features)
```

```python
submission = sample_submission.copy()
```

```python
submission['sales'] = test_pred
```

```python
submission.to_csv('xgboost_submission.csv', index=False)
```

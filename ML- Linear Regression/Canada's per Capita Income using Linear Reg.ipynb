```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income in USD')
plt.scatter(df.year, df.income)
```

```python
reg= linear_model.LinearRegression()
```

```python
reg.fit(df[['year']], df.income)
```

```python
reg.predict([[2020]])
```

```python
reg.predict([[2024]])
```

```python
reg.predict([[2050]])
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
plt.plot(df.year, reg.predict(df[['year']]), color='blue')
```
